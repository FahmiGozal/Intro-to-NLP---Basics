```python
# Create a function to convert timestamp into seconds
import re
def convert_timestamp_into_seconds(timestamp):
    regex_number_finder = "\d+"
    temp_list = re.findall(regex_number_finder, timestamp)
    seconds = int(temp_list[0])*3600 + int(temp_list[1])*60 + int(temp_list[2])

    return seconds


    
    # your code here
```

Who speaks faster?

Tokenize the text, preprocess the tokens so that you have only words (excluding punctuation) and compute the velocity of each speaker as: number of words per interval / length of interval . Store the result in a column named Velocity and compute the average for each speaker.

```python
# Preprocess the data
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python


# Create a function to remove punctuation from text

def remove_punctuation(text):
    doc = nlp(text)
    token_list = [token for token in doc if not token.is_punct]
    no_punct_sentence = ' '.join(token.text for token in token_list)
    return no_punct_sentence

# Create a function to count the non punctuation token of a text

def count_tokens(text):
    doc = nlp(text)
    token_count = len(doc)
    return token_count
    
# Create a function to remove stop words from text
    
def remove_stopwords(text):
    doc = nlp(text)
    token_list = [token for token in doc if not token.is_punct]
    filtered_list = []
    for token in token_list:
        if token.text in nlp.Defaults.stop_words == False:
            filtered_list.append(token)

    no_stopword_sentence = ' '.join(token.text for token in filtered_list)
    return no_stopword_sentence
            


```

```python
# CALL THE FUNCTIONS

df_copy["TextNoPunct"] = df_copy['Text'].apply(lambda x: remove_punctuation(x))
df_copy["n_tokens"] = df_copy['Text'].apply(lambda x: count_tokens(x))
df_copy["TextNoStopword"] = df_copy['Text'].apply(lambda x: remove_stopwords(x))
```

```python
# Compute the velocity and store it in a new column


df_copy["Velocity"] = df_copy['n_tokens']/df_copy['Interval']
df_copy
```
